```python
# https://leetcode.com/problems/big-countries/ 
import pandas as pd

def big_countries(world: pd.DataFrame) -> pd.DataFrame:
    ans = world.loc[(world['area'] >= 3000000) | (world['population']>=25000000),['name','population','area']]
    return ans
```

```python
# https://leetcode.com/problems/recyclable-and-low-fat-products/submissions/1808410944/ 
import pandas as pd

def find_products(products: pd.DataFrame) -> pd.DataFrame:
    low_fat_and_recyclable = products.loc[(products['low_fats'] == 'Y') & (products['recyclable']=='Y'),['product_id']]
    return low_fat_and_recyclable
    
```

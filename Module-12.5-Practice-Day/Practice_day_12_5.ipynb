```python
# Create a Pandas Series from the list [10, 20, 30, 40, 50]. Print the Series, its index, and data type. Then create a DataFrame with columns "Name", "Age", and "City" containing at least 4 rows of data.

import pandas as pd
lst = [10, 20, 30, 40, 50]
pd_series = pd.Series(lst)
print("Series: ",pd_series)
print("Datatype: ",pd_series.dtype)
print("Index: ",list(pd_series.index))

# Then create a DataFrame with columns "Name", "Age", and "City" containing at least 4 rows of data.
d_frame_data = {
    "Name":["Alis","Bob","Devid","Jack"],
    "Age":[23,44,56,78],
    "City":["Dhaka","Khulna","Magura","Kustia"]
}
df = pd.DataFrame(d_frame_data)
df
```

```python
# Using any DataFrame, write code to print the number of rows and columns, column data types (.dtypes), and summary statistics (.describe()).
d_frame_data = {
    "Name":["Alis","Bob","Devid","Jack"],
    "Age":[23,44,56,78],
    "City":["Dhaka","Khulna","Magura","Kustia"]
}
df = pd.DataFrame(d_frame_data)
rows,columns = df.shape
print(f"Rows: {rows}, Columns: {columns}")
print(f"Column's datatypes: {df.dtypes}")
print(f"Statistics :{df.describe()}")
```

```python
# Create a DataFrame using the dictionary below:

import pandas as pd
data = {
    'Product': ['Pen', 'Book', 'Eraser', 'Pencil'],
    'Price': [10, 50, 5, 7],
    'Quantity': [100, 60, 200, 150]
}
df_2 = pd.DataFrame(data)
print(f"Columns : {list(df_2.columns)}")
print(f"Rows : {list(df_2.index)}")
print(f"Single Column : {list(df_2['Product'])}")
print(f"Single Row : {df_2.iloc[0]}")
# df_2
```
